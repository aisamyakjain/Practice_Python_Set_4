```python
# Q2. Write a Python class called "Rectangle" with attributes for "width" and "height". Implement methods 
# to calculate the area and perimeter of the rectangle. Create an instance of this class and use it to print out 
# the rectangle's area and perimeter

class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def calculate_area(self):
        return self.width * self.height

    def calculate_perimeter(self):
        return 2 * (self.width + self.height)


# Create an instance of the Rectangle class
rectangle = Rectangle(4, 6)

# Calculate and print the area
area = rectangle.calculate_area()
print("Area:", area)

# Calculate and print the perimeter
perimeter = rectangle.calculate_perimeter()
print("Perimeter:", perimeter)
```

```python
# Q3. Write a Python class called "BankAccount" with attributes for "balance" and "interest_rate". 
# Implement methods to deposit and withdraw money from the account, as well as to calculate the interest 
# earned on the balance based on the interest rate. Create an instance of this class and use it to test out the 
# implemented methods for deposit, withdrawal, and interest calculation

class BankAccount:
    def __init__(self, balance=0, interest_rate=0):
        self.balance = balance
        self.interest_rate = interest_rate

    def deposit(self, amount):
        self.balance += amount
        print("Deposited:", amount)

    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance -= amount
            print("Withdrawn:", amount)
        else:
            print("Insufficient funds.")

    def calculate_interest(self):
        interest = self.balance * (self.interest_rate / 100)
        return interest

# Create an instance of the BankAccount class
account = BankAccount(balance=1000, interest_rate=2)

# Deposit money into the account
account.deposit(500)

# Withdraw money from the account
account.withdraw(200)

# Calculate and print the interest earned
interest_earned = account.calculate_interest()
print("Interest Earned:", interest_earned)


        
        
```

```python
# Q4. Write a Python class called "Animal" with attributes for "name" and "species". Create a subclass of 
# "Animal" called "Dog" with additional attributes for "breed" and "owner". Implement both classes with 
# appropriate methods and constructors to initialise their properties. Create instances of both classes and 
# use them to print out the various properties of the animals and dogs, such as their names, species, 
# breeds, and owners





class Animal:
    def __init__(self, name, species):
        self.name = name
        self.species = species

    def get_name(self):
        return self.name

    def get_species(self):
        return self.species


class Dog(Animal):
    def __init__(self, name, species, breed, owner):
        super().__init__(name, species)
        self.breed = breed
        self.owner = owner

    def get_breed(self):
        return self.breed

    def get_owner(self):
        return self.owner


# Create an instance of the Animal class
animal = Animal("Max", "Mammal")

# Create an instance of the Dog class
dog = Dog("Buddy", "Canine", "Labrador Retriever", "John")

# Print properties of the animal
print("Animal:")
print("Name:", animal.get_name())
print("Species:", animal.get_species())

# Print properties of the dog
print("\nDog:")
print("Name:", dog.get_name())
print("Species:", dog.get_species())
print("Breed:", dog.get_breed())
print("Owner:", dog.get_owner())
```

```python
# Q7. Create a class called "Shape" with methods to calculate the area and perimeter of the shape. Create 
# subclasses of "Shape" for "Rectangle", "Circle", and "Triangle" with their own methods for calculating area 
# and perimeter. Create objects of each class and print out their area and perimeter

import math

class Shape:
    def calculate_area(self):
        pass

    def calculate_perimeter(self):
        pass


class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def calculate_area(self):
        return self.width * self.height

    def calculate_perimeter(self):
        return 2 * (self.width + self.height)


class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def calculate_area(self):
        return math.pi * self.radius ** 2

    def calculate_perimeter(self):
        return 2 * math.pi * self.radius


class Triangle(Shape):
    def __init__(self, side1, side2, side3):
        self.side1 = side1
        self.side2 = side2
        self.side3 = side3

    def calculate_area(self):
        # Using Heron's formula to calculate the area of a triangle
        s = (self.side1 + self.side2 + self.side3) / 2
        return math.sqrt(s * (s - self.side1) * (s - self.side2) * (s - self.side3))

    def calculate_perimeter(self):
        return self.side1 + self.side2 + self.side3


# Create objects of each shape and print their area and perimeter

rectangle = Rectangle(4, 6)
print("Rectangle:")
print("Area:", rectangle.calculate_area())
print("Perimeter:", rectangle.calculate_perimeter())

print()

circle = Circle(5)
print("Circle:")
print("Area:", circle.calculate_area())
print("Perimeter:", circle.calculate_perimeter())

print()

triangle = Triangle(3, 4, 5)
print("Triangle:")
print("Area:", triangle.calculate_area())
print("Perimeter:", triangle.calculate_perimeter())
```

```python
# Q8. Create a class called "Employee" with properties for "name", "id", and "salary". Add methods to give 
# the employee a raise and to calculate their annual salary. Create objects of this class and test out the 
# methods

class Employee:
    def __init__(self, name, employee_id, salary):
        self.name = name
        self.employee_id = employee_id
        self.salary = salary

    def give_raise(self, amount):
        self.salary += amount

    def calculate_annual_salary(self):
        return self.salary * 12


# Create objects of the Employee class and test the methods

employee1 = Employee("John Doe", 12345, 5000)
print("Employee 1:")
print("Name:", employee1.name)
print("Employee ID:", employee1.employee_id)
print("Salary:", employee1.salary)

print()

employee2 = Employee("Jane Smith", 67890, 7000)
print("Employee 2:")
print("Name:", employee2.name)
print("Employee ID:", employee2.employee_id)
print("Salary:", employee2.salary)

print()

# Give a raise to employee1
employee1.give_raise(1000)
print("Employee 1 (after raise):")
print("Name:", employee1.name)
print("Employee ID:", employee1.employee_id)
print("Salary:", employee1.salary)

print()

# Calculate the annual salary of employee2
annual_salary = employee2.calculate_annual_salary()
print("Employee 2 (annual salary):", annual_salary)
```

```python
# Q9. Create a class called "Book" with properties for "title", "author", and "publisher". Create a subclass of 
# "Book" called "Ebook" with additional properties for "file_format" and "file_size". Create objects of both 
# classes and print out their properties

class Book:
    def __init__(self, title, author, publisher):
        self.title = title
        self.author = author
        self.publisher = publisher

    def get_title(self):
        return self.title

    def get_author(self):
        return self.author

    def get_publisher(self):
        return self.publisher


class Ebook(Book):
    def __init__(self, title, author, publisher, file_format, file_size):
        super().__init__(title, author, publisher)
        self.file_format = file_format
        self.file_size = file_size

    def get_file_format(self):
        return self.file_format

    def get_file_size(self):
        return self.file_size


# Create an instance of the Book class
book = Book("The Great Gatsby", "F. Scott Fitzgerald", "Charles Scribner's Sons")

# Create an instance of the Ebook class
ebook = Ebook("The Catcher in the Rye", "J.D. Salinger", "Little, Brown and Company", "PDF", "2.5 MB")

# Print properties of the book
print("Book:")
print("Title:", book.get_title())
print("Author:", book.get_author())
print("Publisher:", book.get_publisher())

# Print properties of the ebook
print("\nEbook:")
print("Title:", ebook.get_title())
print("Author:", ebook.get_author())
print("Publisher:", ebook.get_publisher())
print("File Format:", ebook.get_file_format())
print("File Size:", ebook.get_file_size())
```

```python
# Q10. Create a class called "Bank" with properties for "name" and "accounts". Add methods to create new 
# accounts, deposit and withdraw money from accounts, and to calculate the total balance of all accounts. 
# Create an object of this class and test out the methods


class Bank:
    def __init__(self, name):
        self.name = name
        self.accounts = []

    def create_account(self, account_name, initial_balance):
        account = {"name": account_name, "balance": initial_balance}
        self.accounts.append(account)

    def deposit(self, account_name, amount):
        for account in self.accounts:
            if account["name"] == account_name:
                account["balance"] += amount
                break

    def withdraw(self, account_name, amount):
        for account in self.accounts:
            if account["name"] == account_name:
                if account["balance"] >= amount:
                    account["balance"] -= amount
                    break
                else:
                    print("Insufficient balance.")
                    break

    def calculate_total_balance(self):
        total_balance = 0
        for account in self.accounts:
            total_balance += account["balance"]
        return total_balance


# Create an object of the Bank class and test the methods

bank = Bank("MyBank")

# Create new accounts
bank.create_account("John Doe", 5000)
bank.create_account("Jane Smith", 8000)

# Deposit money
bank.deposit("John Doe", 2000)
bank.deposit("Jane Smith", 3000)

# Withdraw money
bank.withdraw("John Doe", 1500)
bank.withdraw("Jane Smith", 2000)

# Calculate total balance
total_balance = bank.calculate_total_balance()

# Print account details and total balance
print("Account Details:")
for account in bank.accounts:
    print("Account Name:", account["name"])
    print("Balance:", account["balance"])
    print()

print("Total Balance:", total_balance)
```
